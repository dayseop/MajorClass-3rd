## EDA for label
* pie와 bar 그리기

```python
labels = df['target']
```

```python
labels.value_counts()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.pie(labels.value_counts(), explode = [0.0, 0.1], autopct = '%1.1f%%', labels = [1,0])
plt.title('Pie plot - Heart Disease')

plt.subplot(1,2,2)
plt.bar(x = [0,1], height = labels.value_counts()/1025, color = ['b', 'orange'], label = [1,0])
plt.legend()
plt.show()
```

## EDA : age vs label
* countplot 그리기
* 성별에 따라 확실한 차이가 보인다

```python
plt.figure(figsize=(12,4))
sns.countplot(x = 'age', hue = 'target', data = df)
plt.show()
```

```python
df[['age', 'target']].groupby('age').value_counts()
```

```python
plt.figure(figsize=(4,4))
fig = sns.countplot(x = 'sex', hue = 'target', data = df)
plt.legend(['No', 'Yes'])
plt.xlabel('sex[0: Female  1 : Male]')
plt.show()
```

```python
pd.crosstab(df['sex'], df['target'], margins = True) ## 일하다보면 서버에서 그래픽을 못 그려서 이렇게 수치로 확인함
```

## EDA : CP vs label
* 심장 통증과 심장병은 연관이 커 보임

```python
plt.figure(figsize=(4,4))
sns.countplot(x = 'cp', hue = 'target', data = df)
plt.legend(['No', 'Yes'])
plt.xlabel('Chest Pain [0, 1, 2, 3]')
plt.show()
```

## EDA: trestbps VS label
* 심장병이 있는 사람과 없는 사람의 혈압은 관계가 없음

```python
plt.figure(figsize = (8,4))
sns.kdeplot(df[df['target']==0]['trestbps'])
sns.kdeplot(df[df['target']==1]['trestbps'])
plt.legend(['target=No', 'target=Yes'])
plt.show()
```

```python
plt.boxplot([df[df['target']==0]['trestbps'], df[df['target']==1]['trestbps']])
plt.xlabel('1: target=No,     2: target=Yes')
```

## EDA: chol vs label
* 심장병과 콜레스테롤과의 관계는 없어보임
* 좀 더 확실하게 알아보려면 유의관계: p-value, 카이제곱

```python
plt.figure(figsize = (8,4))
sns.kdeplot(df[df['target']==0]['chol'])
sns.kdeplot(df[df['target']==1]['chol'])
plt.legend(['target=No', 'target=Yes'])
plt.show()
```

## EDA: restecg(심전도: 0, 1, 2)
* 카이제곱 써보기

```python
fig, ax = plt.subplots(1, 1, figsize=(4,4))
sns.countplot(x = df['restecg'], hue=df['target'], data = df)
```

```python
df_chi = pd.DataFrame({'ecg0' : [283, 214], 'ecg1' :[204, 309], 'ecg2' : [12, 3]})

from scipy.stats import chi2_contingency

chi, p, dof, ef = chi2_contingency(df_chi)

print('chi square: {}'.format(chi))
print('P-value: {}'.format(p))
print(dof)
```
