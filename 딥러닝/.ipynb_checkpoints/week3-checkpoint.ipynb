```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.array([10, 7, 5, 4])
result = np.array([1, 1, 0, 0])
score = np.array([1, 1, 0, 0])
plt.scatter(hours, result)
plt.xlim(0, 15)
plt.ylim(-0.1, 1.1)
```

```python
def LinR(x, w = 2, b = -10):
  return w* x + b
print(LinR(hours))
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))

def LogR(x, w=2, b=-10):
  z = LinR(x, w=w, b=b)
  h_z = sigmoid(z)
  return h_z


print(LogR(hours))
```

```python
for i in range(1, 100, 1):
  w = i/10
  b = 0
  x = np.arange(-40,46,1)
  h_z = LogR(x, w=w, b=b)
  plt.plot(x, h_z)

plt.show()
```

```python
for i in range(-40, 41, 20):
  w = 1
  b = i
  x = np.arange(-60, 66, 1)
  h_z = LogR(x, w=w, b=b)
  plt.plot(x, h_z, label = 'bias:'+str(b))

plt.legend()
plt.show() ## 왼쪽으로 이동 ++
```

```python
class LR():
  def __init__(self, w=2, b=-10):
    self.w = w
    self.b = b

  def LinR(self, x):
    return self.w * x + self.b

  def sigmoid(self, z):
    return 1/(1+np.exp(-z))

  def LogR(self, x):
    z = self.LinR(x)
    h_z = self.sigmoid(z)
    return h_z

model = LR()

print(model.LogR(hours))
```

```python
# plot model

x = np.linspace(-1, 12, 100) ## -1 ,12 를 100개 배열로 균등 분할
pred = model.LogR(x)
plt.plot(x, pred)
plt.scatter(hours, result)
```

```python
# prediction 학습안한거 한개 오류
pred = model.LogR(hours)
print(pred)
pred[pred>=0.5] = 1.0; pred[pred<0.5] = 0.0
print(pred)
```

```python
#cost

def cost_function(h, y):
  return -(y*np.log(h)+(1-y)*np.log(1-h))

pred = model.LogR(hours)
print(cost_function(pred, np.array(result)))
```

```python
print(f'Gradient: {pred - score}')
print(f'Mean: {np.mean(pred - score)}')

lr = 2.0
w = 2; b = -10

new_b = b - lr*np.mean(pred-score)
print(f'new bias: {new_b}')
```

```python
print(f'old model: {LogR(hours)}')
print(f'new model: {LogR(hours, b = -10.3)}')
```

```python
x = np.linspace(-1, 12, 100)

# old
model.b = -10.0
pred = model.LogR(x)
plt.plot(x, pred, label='old')

# updata
model.b = -10.3
pred = model.LogR(x)
plt.plot(x,pred, label = 'New')

plt.scatter(hours, score)

plt.legend()
```

```python
class LR():
  def __init__(self, w=2, b=-10, lr = 0.01):
    self.w = w
    self.b = b
    self.lr = lr

  def LinR(self, x):
    return self.w * x + self.b

  def sigmoid(self, z):
    return 1/(1+np.exp(-z))

  def LogR(self, x):
    z = self.LinR(x)
    h_z = self.sigmoid(z)
    return h_z

  def fit(self, x, y, epochs = 1000):
    for epoch in range(epochs):
      pred = self.LogR(x)
      self.w = np.mean(self.w - self.lr*np.mean(pred-y)*x)
      self.b = np.mean(self.b - self.lr*np.mean([pred-y]))
model = LR()
model.fit(hours, score)

print(f'weight: {model.w}')
print(f'bias: {model.b}')
```

```python
#plot model

x = np.linspace(-1, 12, 100)

# old
old_model = LR()
pred = old_model.LogR(x)
plt.plot(x, pred, label = 'old model')

# new
pred = model.LogR(x)
plt.plot(x, pred, label='new model')
plt.legend()

plt.scatter(hours, score)

pred = model.LogR(hours)
pred[pred>=0.5] = 1.0; pred[pred<0.5]=0.0
print(f'pred: {pred}')
```
